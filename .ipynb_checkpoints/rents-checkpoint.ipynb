```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set_style('whitegrid')
```

```python
#importing dataset
df = pd.read_csv('nigeria-rent.csv')
#making a copy of the dataframe
dfc = df.copy()
dfc
```

```python
#info on each column of the dataframe
dfc.info()
```

```python
#dropping duplicates in the data
dfc.duplicated().sum()
dfc.drop_duplicates(inplace=True)
dfc.duplicated().sum()
```

```python
#treating null values
dfc.isnull().sum()
dfc.dropna(inplace=True)
dfc.isnull().sum()
```

```python
#dropping unnecessary more info column
#dfc.drop('More Info', axis = 1, inplace=True)
dfc.head(10)
```

```python
#splitting the location column to get the states and city
loc_split = dfc['Location'].str.split()
state = [state[-1] for state in loc_split]
city = list(loc_split.apply(lambda x: x[-2] if len(x) >= 2 else None))
#replacing the location column value with the state values
dfc['Location'] = state
#creating a city column in the dataframe
dfc['City'] = city
```

```python
#cleaning the price column
#splitting the values from the / symbol to seperate the text and numbers
split_price = dfc['Price'].str.split('/')
#creating a list comprehension of the first values of each row
new_price = [price[0] for price in split_price]
dfc['Price'] = new_price
dfc['Price'] = dfc['Price'].str.replace(',', '')
dfc['Price'] = dfc['Price'].astype(np.int64)
```

```python
d_ind = dfc[dfc['Bedrooms'] == ' beds'].index
dfc.loc[d_ind, 'Bedrooms'] = dfc.loc[d_ind, 'Bedrooms'].str.replace(' beds', '0 beds')
dfc['Bedrooms'] = dfc['Bedrooms'].str.replace(' beds', '')
dfc['Bedrooms'] = dfc['Bedrooms'].astype(np.int64)
```

```python
d_ind = dfc[dfc['Bathrooms'] == ' baths'].index
dfc.loc[d_ind, 'Bathrooms'] = dfc.loc[d_ind, 'Bathrooms'].str.replace(' baths', '0 baths')
dfc['Bathrooms'] = dfc['Bathrooms'].str.replace(' baths', '')
dfc['Bathrooms'] = dfc['Bathrooms'].astype(np.int64)
```

```python
d_ind = dfc[dfc['Toilets'] == ' Toilets'].index
dfc.loc[d_ind, 'Toilets'] = dfc.loc[d_ind, 'Toilets'].str.replace(' Toilets', '0 Toilets')
dfc['Toilets'] = dfc['Toilets'].str.replace(' Toilets', '')
dfc['Toilets'] = dfc['Toilets'].astype(np.int64)
```

```python
#visualizing the city with the most buildings
loc = dfc['City'].value_counts().sort_values(ascending=False).reset_index().head(5)
sns.barplot(data=loc, x='City', y='count', palette='magma')
plt.show()
```

The above graph shows a comparison of the number of properties across the cities within the top 5 states: Lagos, Oyo, Abuja, Rivers, Ogun. From the chart, Lagos dominated the most with more properties than the other states combined and Ibadan being the only city from Oyo in this chart. 
The vertical bars represent each city and their height corresponding to the number of properties, insights from the chart above includes:

* Lekki, which is located in Lagos state, has the highest number of properties by a large margin significantly beating other cities listed which could be due to several factors and demographics.
* Ajah located in Lagos state and Ibadan located in Oyo state follow Lekki, with Ajah surpassing Ibadan with a slight margin and both significantly lesser than Lekki.
* Overall, the distribution of number of properties built across each state is not uniform as we can see cities within Lagos being the favourite location to build.

```python
#Filtering for the top 5 states with the highest number of buildings
top_5 = dfc['Location'].value_counts().head(5).index
top_5_data = dfc[dfc['Location'].isin(top_5)]
#Grouping the filtered data by the city and the stte of the building
grouped = top_5_data.groupby(['City', 'Newly Built']).size().sort_values(ascending = False).reset_index(name='Number of Properties').head(6)
#Visualizing the number of new and old buildings per city
plt.figure(figsize=(8,6))
sns.barplot(data=grouped, x='Newly Built', y='Number of Properties', hue='City', palette='magma')
plt.xticks([1,0],['Yes', 'No'])
plt.title('Number of Old and New Buildings per City')
plt.legend(loc='right')
plt.show()
```

From the chart above, Lekki has the highest number of old buildings via a large margin compared to the other cities. For buildings that are newly built, Lekki also has the highest number but it can't be compared to the number of old buildings located in the city. 


```python
#distribution of properties by number of bedrooms
bedroom_dist = dfc['Bedrooms'].value_counts()
#visualizing the distribution
plt.figure(figsize=(8,6))
sns.barplot(x=bedroom_dist.index, y=bedroom_dist.values, palette='rocket')
plt.title('Number of Bedroom Distribution')
plt.xlabel('Number of Bedrooms')
plt.tight_layout()
plt.show()
```

From the chart above, buildings with 3 bedrooms are the most popular or preferred type of buildings across states in the dataset, while there is an extremely low preferrability for building with 9 bedrooms. Following the 3 bedrooms building are those with 2 and 1 bedrooms showing a good amount of favorability in the property market.

```python
plt.figure(figsize=(8,6))
sns.kdeplot(data=top_5_data, x='Bedrooms', color = 'b', label='Bedrooms')
sns.kdeplot(data=top_5_data, x='Bathrooms', fill = True, color='g', label='Bathrooms')
plt.xlabel('Number of Bedrooms/Bathrooms')
plt.title('Distribution of Number of Bathrooms per Number of Bedrooms')
plt.legend()
plt.show()
```

There is a steady increase in the number of buildings as the number of bedrooms increase seeing it peak at 3 bedrooms and then a rapid fall off in the number of buildings having bedrooms that are more than 3.
In the case of the number of bathrooms, buildings without bathrooms outnumber those with 1 or more bathrooms and bedrooms which goes to show that a good ratio of buildings were erected without much consideration for the bathrooms.
